   # Project: Stroke Prediction Based on Input Parameters

According to the World Health Organization (WHO) stroke is the 2nd leading cause of death globally, responsible for approximately 11% of total deaths. This dataset is used to predict whether a patient is likely to get stroke based on the input parameters like gender, age, various diseases, and smoking status. Each row in the data provides relavant information about the patient. Attribute Information:

    1. id: unique identifier
    2. gender: "Male", "Female" or "Other"
    3. age: age of the patient
    4. hypertension: 0 if the patient doesn't have hypertension, 1 if the patient has hypertension
    5. heart_disease: 0 if the patient doesn't have any heart diseases, 1 if the patient has a heart disease
    6. ever_married: "No" or "Yes"
    7. work_type: "children", "Govt_jov", "Never_worked", "Private" or "Self-employed"
    8. Residence_type: "Rural" or "Urban"
    9. avg_glucose_level: average glucose level in blood
    10.bmi: body mass index
    11.smoking_status: "formerly smoked", "never smoked", "smokes" or "Unknown"*
    12.stroke: 1 if the patient had a stroke or 0 if not
    *Note: "Unknown" in smoking_status means that the information is unavailable for this patient

```python
Row_list =[]
  
# Iterate over each row
for index, rows in stroke.iterrows():
    # Create list for the current row
    my_list =[rows.age]
      
    # append the list to the final list
    Row_list.append(my_list)
  
# Print the list
print(Row_list)
```

```python
Row_list2 =[]
  
# Iterate over each row
for index, rows in stroke.iterrows():
    # Create list for the current row
    my_list2 =[rows.avg_glucose_level]
      
    # append the list to the final list
    Row_list2.append(my_list2)
  
# Print the list
print(Row_list2)
```

```python
Row_list3 =[]
  
# Iterate over each row
for index, rows in stroke.iterrows():
    # Create list for the current row
    my_list3 =[rows.bmi]
      
    # append the list to the final list
    Row_list3.append(my_list3)
  
# Print the list
print(Row_list3)
```

```python
class Patients:
    
    def __init__(self,identity):
        self.identity = identity
        print("The patient identification number is", self.identity)
    
    def min_age(self):
        return min(Row_list)
    
    def max_age(self):
        return max(Row_list)
    
    
```

```python
test = Patients(1234)
print("Minimum Age:", test.min_age())
print("Maximum Age:", test.max_age())
```

```python
class average_glucose_level(Patients):
        
    def min_avg_glucose_level(self):
        return min(Row_list2)
    
    def max_avg_glucose_level(self):
        return max(Row_list2)
    
```

```python
test2 = average_glucose_level(9046)
print("Minimum Average of glucose level:", test2.min_avg_glucose_level())
print("Maximum Average of glucose level:", test2.max_avg_glucose_level())
```

```python
class patient_bodymassindex(Patients):
        
    def min_bmi(self):
        return min(Row_list3)
    
    def max_bmi(self):
        return max(Row_list3)
    
```

```python
test3 = patient_bodymassindex(9046)
print("Minimum Body Mass Index:", test3.min_bmi())
print("Maximum Body Mass Index:", test3.max_bmi())
```
